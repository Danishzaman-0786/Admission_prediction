```python
#we know that scaling only aply on training data_set:
#then first we split the Data by train_test_split called by sklearn:
from sklearn.model_selection import train_test_split as tts
a_train,a_test,b_train,b_test=tts(x,y,test_size=0.2,random_state=1)
a_train
```

```python
#we will use minmax scaling:
#formula (min_max)=>[xi â€² = xi âˆ’ min ( x ) /max ( x ) âˆ’ min ( x )]:
#minmax scaling used only for upper limit(boundary):
#Zscore used only for upper and lower limit(boundary):
#we also now tha scaling aplied only on input columns:
#here is used minmax_scaling:
from sklearn.preprocessing import MinMaxScaler as mms
#we make a class of MinMaxScaler:
mms=mms()
```

```python
a1_train=mms.fit_transform(a_train)
a1_test=mms.transform(a_test)
```

```python
a1_train.shape
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
#now we make model by Neurons:
model=Sequential()
```

```python
#we make a layers like input layes  and output layers
# input layers=>model.add(Dense(7,activation="relu")):
model.add(Dense(7,activation="relu",input_dim=7))
model.add(Dense(1,activation="relu"))
#output layer =>mdel.add(Dense(1,activation="sigmoid")):
model.add(Dense(1,activation="linear"))
```

```python
model.summary()
```

```python
model.compile(loss="mean_squared_error",optimizer="Adam")
```

```python
history=model.fit(a1_train,b_train,epochs=100,validation_split=0.2)
```
